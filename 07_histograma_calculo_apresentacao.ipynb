#### HISTOGRAMA - CÁLCULO E APRESENTAÇÃO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calcular_histograma(imagem):

    # Inicializando um vetor para armazenar as frequências de intensidade
    histograma = np.zeros(256, dtype=int)

    # Percorrendo todos os pixels da imagem
    altura, largura = imagem.shape
    for i in range(altura):
        for j in range(largura):
            # Obtendo o valor de intensidade do pixel atual
            intensidade = int(imagem[i, j])
            # Incrementando a contagem de frequência para esse valor de intensidade
            histograma[intensidade] += 1

    return histograma
```

```python
# Carregando a imagem em escala de cinza
imagem = cv2.imread('Imagens/00_edilson.png', cv2.IMREAD_GRAYSCALE)

# Calculando o histograma
histograma = calcular_histograma(imagem)

# Plotando o histograma
plt.plot(histograma, color='black')
plt.xlabel('Intensidade')
plt.ylabel('Frequência')
plt.title('Histograma da Imagem em Escala de Cinza')
plt.show()
```
