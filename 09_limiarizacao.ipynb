#### LIMIARIZAÇÃO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregando a imagem em escala de cinza
imagem = cv2.imread('Imagens/00_edilson.png', cv2.IMREAD_GRAYSCALE)

# Definindo um valor de limiar
limiar = 127

# Obtendo as dimensões da imagem
altura, largura = imagem.shape

# Inciando a imagem limiarizada
imagem_limiarizada = np.zeros_like(imagem)

# Percorrendo a imagem e aplicando a limiarização manualmente
for i in range(altura):
    for j in range(largura):
        # Obtendo o valor do pixel na posição (i, j)
        valor_pixel = imagem[i, j]

        # Aplicando a limiarização
        if valor_pixel < limiar:
            imagem_limiarizada[i, j] = 0  # Pixel preto
        else:
            imagem_limiarizada[i, j] = 255  # Pixel branco
```

```python
# Salvando a imagem limiarizada em um arquivo
cv2.imwrite('Imagens/09_edilson_limiarizada.png', imagem_limiarizada)

# Mostrando a imagem limiarizada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_limiarizada, cmap='gray')
plt.title('Imagem Limiarizada')
plt.show()
```
