#### 10 - MULTILIMIARIZAÇÃO

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def multilimiarizacao(imagem, limiares):
    # Inicializando a imagem multilimiarizada
    imagem_multilimiarizada = np.zeros_like(imagem)

    # Percorrendo os limiares
    for i in range(len(limiares) - 1):
        # Segmentar a imagem entre os limiares
        mascara = cv2.inRange(imagem, limiares[i], limiares[i+1])

        # Atribuindo um valor específico para cada segmento
        imagem_multilimiarizada[mascara == 255] = i * (255 // (len(limiares) - 1))

    return imagem_multilimiarizada
```

```python
# Carregando a imagem em escala de cinza
imagem = cv2.imread('Imagens/00_edilson.png', cv2.IMREAD_GRAYSCALE)

# Definindo os limiares
limiares = [50, 100, 150, 200]

# Aplicando a multilimiarização
imagem_multilimiarizada = multilimiarizacao(imagem, limiares)

# Salvando a imagem multilimiarizada em um arquivo
cv2.imwrite('Imagens/10_edilson_multilimiarizada.png', imagem_multilimiarizada* 170 // len(limiares))

# Exibindo a imagem multilimiarizada
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(imagem, cmap='gray')
plt.title('Imagem Original')
plt.subplot(1, 2, 2)
plt.imshow(imagem_multilimiarizada, cmap='gray')
plt.title('Imagem Multilimiarizada')
plt.show()
```
